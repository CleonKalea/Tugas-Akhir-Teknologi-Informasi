```python
import pandas as pd
```

```python
df_bangli = pd.read_csv('../Data/STAGE 3 PREPROCESSING/STAGE_3_PREPROCESSING_Bangli_FINAL.csv')
df_gianyar = pd.read_csv('../Data/STAGE 3 PREPROCESSING/STAGE_3_PREPROCESSING_Gianyar_FINAL.csv')
df_klungkung = pd.read_csv('../Data/STAGE 3 PREPROCESSING/STAGE_3_PREPROCESSING_Klungkung_FINAL.csv')
df_tabanan = pd.read_csv('../Data/STAGE 3 PREPROCESSING/STAGE_3_PREPROCESSING_Tabanan_FINAL.csv')
df_jembrana = pd.read_csv('../Data/STAGE 3 PREPROCESSING/STAGE_3_PREPROCESSING_Jembrana_FINAL.csv')
df_karangasem = pd.read_csv('../Data/STAGE 3 PREPROCESSING/STAGE_3_PREPROCESSING_Karangasem_FINAL.csv')
df_singaraja = pd.read_csv('../Data/STAGE 3 PREPROCESSING/STAGE_3_PREPROCESSING_Singaraja_FINAL.csv')
df_jakarta_timur = pd.read_csv('../Data/STAGE 3 PREPROCESSING/STAGE_3_PREPROCESSING_Jakarta_Timur_FINAL.csv')
df_jakarta_utara = pd.read_csv('../Data/STAGE 3 PREPROCESSING/STAGE_3_PREPROCESSING_Jakarta_Utara_FINAL.csv')
```

```python
print("df_bangli:", len(df_bangli))
print("df_gianyar:", len(df_gianyar))
print("df_klungkung:", len(df_klungkung))
print("df_tabanan:", len(df_tabanan))
print("df_jembrana:", len(df_jembrana))
print("df_karangasem:", len(df_karangasem))
print("df_singaraja:", len(df_singaraja))
print("df_jakarta_timur:", len(df_jakarta_timur))
print("df_jakarta_utara:", len(df_jakarta_utara))
```

```python
dfs = [df_bangli, df_gianyar, df_klungkung, df_tabanan, df_jembrana, df_karangasem, df_singaraja, df_jakarta_timur, df_jakarta_utara]

columns = dfs[0].columns
all_same_columns = all(df.columns.equals(columns) for df in dfs)

print("Columns Match:", all_same_columns)
```

```python
df_merged = pd.concat(dfs, ignore_index=True)
```

```python
df_merged
```

```python
unique_klasifikasi_perkara_values = df_merged['klasifikasi_perkara'].unique()
num_unique_klasifikasi_perkara = len(unique_klasifikasi_perkara_values)
unique_klasifikasi_counts = df_merged['klasifikasi_perkara'].value_counts()

print(f"Number of unique values in 'status_perkara': {num_unique_klasifikasi_perkara}")
print(unique_klasifikasi_counts)
```

```python
df_merged.to_csv('../Data/STAGE 4 FINAL MERGED DATA/STAGE_4_MERGED_FINAL.csv', index=False)
```
