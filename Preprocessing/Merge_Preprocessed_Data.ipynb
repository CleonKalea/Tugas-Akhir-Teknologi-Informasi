## Temporary

```python

def leftover_cleaning(text):
    text = re.sub(r'\[.*?\]', '', str(text))
    text = re.sub(r'\s*\(\s*(nol koma|satu|dua|tiga|empat|lima|enam|tujuh|delapan|sembilan|sepuluh|sebelas|dua belas|tiga belas|empat belas|lima belas|enam belas|tujuh belas|delapan belas|sembilan belas|dua puluh)\s*\)\s*', ' ', text)
    text = re.sub(r'\*\*', '', text)
    text = re.sub(r';', ',', text)

    return text

df_merged['cleaned_barang_bukti'] = df_merged['cleaned_barang_bukti'].apply(leftover_cleaning)
```
