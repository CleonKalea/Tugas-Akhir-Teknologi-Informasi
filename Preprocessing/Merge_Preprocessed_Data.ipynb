# Map Encoding

## Encoding: Klasifikasi Perkara

```python
klasifikasi_perkara_unique_names = df_selected['klasifikasi_perkara'].unique()
klasifikasi_perkara_mapping = {name: i for i, name in enumerate(klasifikasi_perkara_unique_names)}

df_selected['klasifikasi_perkara_encoded'] = df_selected['klasifikasi_perkara'].map(klasifikasi_perkara_mapping)

with open('../Mapping/klasifikasi_perkara_mapping.pkl', 'wb') as f:
    pickle.dump(klasifikasi_perkara_mapping, f)
    
for name, code in klasifikasi_perkara_mapping.items():
    print(f"{code} <- {name} ")
```

## Encoding: Penuntut Umum

```python
penuntut_umum_unique_names = df_selected['penuntut_umum'].unique()
penuntut_umum_mapping = {name: i for i, name in enumerate(penuntut_umum_unique_names)}

df_selected['penuntut_umum_encoded'] = df_selected['penuntut_umum'].map(penuntut_umum_mapping)

with open('../Mapping/penuntut_umum_mapping.pkl', 'wb') as f:
    pickle.dump(penuntut_umum_mapping, f)
    
for name, code in penuntut_umum_mapping.items():
    print(f"{code} <- {name} ")
```

## Encoding: Hakim

```python
hakim_unique_names = df_selected['hakim'].unique()
hakim_mapping = {name: i for i, name in enumerate(hakim_unique_names)}

df_selected['hakim_encoded'] = df_selected['hakim'].map(hakim_mapping)

with open('../Mapping/hakim_mapping.pkl', 'wb') as f:
    pickle.dump(hakim_mapping, f)
    
for name, code in hakim_mapping.items():
    print(f"{code} <- {name} ")
```

## Load Saved Mapping

```python
with open('../Mapping/klasifikasi_perkara_mapping.pkl', 'rb') as f:
    klasifikasi_perkara_mapping = pickle.load(f)

print(klasifikasi_perkara_mapping)

with open('../Mapping/penuntut_umum_mapping.pkl', 'rb') as f:
    penuntut_umum_mapping = pickle.load(f)

print(penuntut_umum_mapping)

with open('../Mapping/hakim_mapping.pkl', 'rb') as f:
    hakim_mapping = pickle.load(f)

print(hakim_mapping)
```
